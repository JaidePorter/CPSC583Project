```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df1 = pd.read_csv('data.csv')
df1.head(3)

df = pd.DataFrame()
df = df1
df.head(3)
```

```python
print(len(df))
df.dropna()
print(len(df))
```

```python
# Initialize the scaler
scaler = MinMaxScaler()

def normalizeValues(colName):
    # Normalize the values in the column to range between 0 and 1
    colName = scaler.fit_transform(df[[colName]])
    return colName
```

```python
df['popularity'] = normalizeValues('popularity').round(3)
df.head(3)
```

```python
df['tempo'] = normalizeValues('tempo').round(3) 
df.head(3)
```

```python
df['duration_ms_normalized'] = normalizeValues('duration_ms').round(3) 
df.head(3)
```

```python
# originally negative value. After normalization it changed every value between 0-1. 
df['loudness'] = normalizeValues('loudness').round(3) 
df.head(3)
```

```python
print(df['energy'].min(), df['energy'].max())
print(df['speechiness'].min(), df['speechiness'].max())
print(df['acousticness'].min(), df['acousticness'].max())
print(df['liveness'].min(), df['liveness'].max())
print(df['valence'].min(), df['valence'].max())
print(df['instrumentalness'].min(), df['instrumentalness'].max())
```

```python
df['energy'] = df['energy'].round(3)
df['speechiness'] = df['speechiness'].round(3)
df['acousticness'] = df['acousticness'].round(3)
df['liveness'] = df['liveness'].round(3)
df['valence'] = df['valence'].round(3)
df['instrumentalness'] = df['instrumentalness'].round(3)
df.head(10)
```

```python
UniqueGenreCounts = df['genre'].value_counts()
print("Total # of genres\n", UniqueGenreCounts)
```

```python
UniqueGenreCounts = df['track_name'].value_counts()
print("Total # of track_name\n", UniqueGenreCounts)
```

```python
UniqueGenreCounts = df['artist_name'].value_counts()
print("Total # of artist_name\n", UniqueGenreCounts)
```
